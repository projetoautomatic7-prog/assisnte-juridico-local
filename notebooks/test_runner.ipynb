```python
import subprocess
import os

def run_command(command):
    """Executa um comando shell e imprime a sa√≠da em tempo real."""
    print(f"üöÄ Executando: {command}\n")
    process = subprocess.Popen(
        command,
        shell=True,
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True,
        cwd=os.path.join(os.getcwd(), "..") # Executa na raiz do projeto
    )

    # Ler sa√≠da linha a linha
    while True:
        output = process.stdout.readline()
        if output == '' and process.poll() is not None:
            break
        if output:
            print(output.strip())

    rc = process.poll()
    return rc
```

```python
run_command("npm run test:run")
```
